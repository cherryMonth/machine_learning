```python
import pandas  as pd
import requests
import time

data = pd.read_csv('data.csv')

data["area"] = data["area"].apply(lambda x: x.replace("[", "").replace("]", ""))
data["province"] = data["area"].apply(lambda x: x.split("")[0])  # 
data["city"] = data["area"].apply(lambda x: x.split("")[1])  # 

j = list()
w = list()

def geocodeG(address):
    par = {'address': address, 'key': '40913b5c086c2ef82192bbd0afff78b7'}
    base = 'http://restapi.amap.com/v3/geocode/geo'
    response = requests.get(base, par)
    answer = response.json()
    result = answer['geocodes']
    return result[0]['location'].split(",") if result else tuple()

length = len(data)

for add in data.itertuples():
    result = geocodeG(add.name) or geocodeG(add.city) or geocodeG(add.province)
    j.append(result[0])
    w.append(result[1])
    print((add.Index +1) * 1.0 / length)
    
```

```python
j = pd.Series(j)
```

```python
w = pd.Series(w)
```

```python
data['longitude'] = j
data['latitude'] = w
```
